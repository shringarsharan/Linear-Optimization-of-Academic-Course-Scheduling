```python
import pandas as pd
from datetime import datetime
import numpy as np
```

# course.csv

```python
schedule = pd.read_csv('schedule.csv')
schedule.head()
```

```python
# Drop records without a course name and select 500-level courses 
schedule.dropna(subset=['course'],axis=0,inplace=True)
course = schedule[(schedule['ID'].str.startswith('5'))&(schedule['term']==20191)]
```

```python
course = course[['course','units','ID','section','first_instructor','first_instructor_uid','first_days',\
                 'first_begin_time','first_end_time','classes_begin','classes_end','seats_offered']]
course.reset_index(drop=True,inplace=True)
```

```python
course
```

```python
# Only schedule courses with 0, 1, 1.5,and 3 credit units
temp = course[(course['units']==1.5)|(course['units']==3)|(course['units']==0)|(course['units']==1)]
# Only schedule courses from Monday to Friday
temp = temp[(temp['first_days']!='S')&(temp['first_days']!='FS')]
# Only schedule courses that are assigned an instructor
temp = temp.dropna(subset=['first_instructor'])
# Only schedule courses that already have a starting date and ending date
temp = temp.dropna(subset=['classes_begin'])
temp = temp.reset_index(drop=True)
```

```python
# all classes start in January and February should be considered starting in the first-half semester 
# all classes start in March should be considered starting in the second-half semester
temp['classes_begin'].unique()
```

```python
temp['classes_end'].unique()
# all classes end in February and March should be considered ending in the first-half semester
# all classes end in April and May should be considered ending in the second-half semester
```

```python
temp['frequency'] = temp["first_days"].apply(lambda x: 2 if x == "MW" or x == "TH" else 1)
```

```python
# Change 0 and 1 unit to 1.5 units 
for i in range(temp.shape[0]):
    if temp.loc[i,'units']==0:
        temp.loc[i,'units']=1.5
    elif temp.loc[i,'units']==1:
        temp.loc[i,'units']=1.5
```

```python
# Decide if this is a first-half, second-half, or full-semeter course
for i in range(temp.shape[0]):
    if temp.loc[i,'classes_begin']<'3/1/2019' and temp.loc[i,'classes_end']<'4/1/2019':
        temp.loc[i,'semester'] = 'first'
    elif temp.loc[i,'classes_begin']<'3/1/2019' and temp.loc[i,'classes_end']>='4/1/2019':
        temp.loc[i,'semester'] = 'full'
    else:
        temp.loc[i,'semester'] = 'second'
```

```python
for i in range(temp.shape[0]):
    if temp.loc[i,'units']==3:
        if temp.loc[i,'semester']=='full':
            if temp.loc[i,'frequency']==1:
                temp.loc[i,'length']=180
            else:
                temp.loc[i,'length']=90
        else:
            if temp.loc[i,'frequency']==1:
                temp.loc[i,'length']=360
            else:
                temp.loc[i,'length']=180
    else:
        if temp.loc[i,'semester']=='full':
            if temp.loc[i,'frequency']==1:
                temp.loc[i,'length']=90
            else:
                temp.loc[i,'length']=45
        else: 
            if temp.loc[i,'frequency']==1:
                temp.loc[i,'length']=180
            else:
                temp.loc[i,'length']=90
# Since all graudate courses have to be either 1.5 hr or 3 hr, classes with 'length'=360 or 45 will be deleted 
```

```python
temp['length'].unique()
```

```python
temp = temp[temp['length']!=45]
```

# faculty.csv

```python
ranks = []
for i in range(temp['first_instructor'].nunique()):
    temp_rank = np.random.choice([1,2,3], size = 3, p=[1/3, 1/3, 1/3], replace = False)
    ranks.append(list(temp_rank))
ranks = pd.DataFrame(ranks,columns=['Morning','Evening','Other'])
ranks['first_instructor'] = temp['first_instructor'].unique()
ranks['timeslot_0'] = ranks['Morning']
ranks['timeslot_1'] = ranks['Morning']
ranks['timeslot_2'] = ranks['Other']
ranks['timeslot_3'] = ranks['Other']
ranks['timeslot_4'] = ranks['Other']
ranks['timeslot_5'] = ranks['Other']
ranks['timeslot_6'] = ranks['Evening']
ranks['timeslot_7'] = ranks['Evening']
ranks['timeslot_8'] = ranks['Evening']
ranks.drop(columns=['Morning','Evening','Other'],inplace=True)
```
